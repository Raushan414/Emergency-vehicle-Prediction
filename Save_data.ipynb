```
import numpy as np
import pandas as pd
import os
import shutil
```

```
orig_dir="/content/drive/My Drive/Colab/Datasets/vehicles_dataset"
temp_dir=os.path.join(orig_dir,'train')
img_dir=os.path.join(temp_dir,'images')

train_df=pd.read_csv(os.path.join(orig_dir,'train.csv'))
test_df=pd.read_csv(os.path.join(orig_dir,'test.csv'))
```

```
#make train,validation,test directory respectively
base_dir="/content/drive/My Drive/Colab/Hackathons/Emergency_vehicle/Data"
os.mkdir(base_dir)
train_dir=os.path.join(base_dir,'train')
os.mkdir(train_dir)
validation_dir=os.path.join(base_dir,'validation')
os.mkdir(validation_dir)
test_dir=os.path.join(base_dir,'test')
os.mkdir(test_dir)

#further split train dir to its classes
train_emergency_dir=os.path.join(train_dir,'emergency')
os.mkdir(train_emergency_dir)
train_non_emergency_dir=os.path.join(train_dir,'non_emergency')
os.mkdir(train_non_emergency_dir)

#further split validation dir to its classes
validation_emergency_dir=os.path.join(validation_dir,'emergency')
os.mkdir(validation_emergency_dir)
validation_non_emergency_dir=os.path.join(validation_dir,'non_emergency')
os.mkdir(validation_non_emergency_dir)
```

```
train_df.head()
```

```
#check for counts of emergency and non_emergency type vehicles
train_df['emergency_or_not'].value_counts()
```

```
#we have 965 non_emergency type vehicle and 681 emergency type vehicle
#split 681 to 581 and 100 to train and validation images and save it to their dir respectively
#split 965 to 800 and 165 to train and validation images and save it to their dir respectively
cnt1=0
cnt2=0
for fname,target in zip(train_df['image_names'],train_df['emergency_or_not']):
  src=os.path.join(img_dir,fname)
  if(target==0):
    if(cnt1<800):
      dst=os.path.join(train_non_emergency_dir,fname)
    else:
      dst=os.path.join(validation_non_emergency_dir,fname)
    cnt1=cnt1+1
  elif(target==1):
    if(cnt2<581):
      dst=os.path.join(train_emergency_dir,fname)
    else:
      dst=os.path.join(validation_emergency_dir,fname)
    cnt2=cnt2+1
  shutil.copyfile(src,dst)
```

```
test_df.head()
```

```
#copy test images to test_dir
for fname in test_df['image_names']:
  src=os.path.join(img_dir,fname)
  dst=os.path.join(test_dir,fname)
  shutil.copyfile(src,dst)
```

```
print('total training emergency images:', len(os.listdir(train_emergency_dir)))
print('total training non_emergency images:', len(os.listdir(train_non_emergency_dir)))
print('total validation emergency images:', len(os.listdir(validation_emergency_dir)))
print('total validation non_emergency images:', len(os.listdir(validation_non_emergency_dir)))
print('total test images:', len(os.listdir(test_dir)))
```
